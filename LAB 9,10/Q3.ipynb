```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, precision_score, recall_score
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
```

```python
dataQ3 = {
    'Height': [150, 155, 160, 165, 170, 175, 180],
    'Weight': [50, 55, 60, 63, 68, 72, 75]
}
dfQ3 = pd.DataFrame(dataQ3)
XQ3 = dfQ3[['Height']]
YQ3 = dfQ3['Weight']
modelQ3 = LinearRegression()
modelQ3.fit(XQ3, YQ3)
print(f"Intercept: {modelQ3.intercept_:.4f}")
print(f"Slope: {modelQ3.coef_[0]:.4f}")
print(f"RÂ²: {modelQ3.score(XQ3, YQ3):.4f}")
prediction172 = modelQ3.predict([[172]])[0]
print(f"Predicted weight for 172 cm height: {prediction172:.2f} kg")
yPredictionQ3 = modelQ3.predict(XQ3)
residualsQ3 = YQ3 - yPredictionQ3
print(f"\nResiduals: {residualsQ3.values}")
print(f"Mean of residuals: {np.mean(residualsQ3):.6f}")
print(f"Std of residuals: {np.std(residualsQ3):.4f}")
print("Linear assumptions appear reasonable: residuals have mean ~0 and relatively constant variance")
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(dfQ3['Height'], dfQ3['Weight'], color='blue', s=100)
plt.plot(dfQ3['Height'], yPredictionQ3, color='red', linewidth=2)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight with Regression Line')
plt.grid(True, alpha=0.3)
plt.subplot(1, 2, 2)
plt.scatter(yPredictionQ3, residualsQ3, color='green', s=100)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
plt.xlabel('Predicted Weight')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('q3HeightWeight.png', dpi=300, bbox_inches='tight')
plt.show()
```

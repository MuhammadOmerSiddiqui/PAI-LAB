```python
import cv2
import numpy as np
image = cv2.imread("image.png")
image2 = np.zeros((300,800,3),dtype=np.uint8)
targetWidth = 600
targetHeight = 800
resizedImage = cv2.resize(image,(targetWidth,targetHeight))
text = "This is a Book On Elon"
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(resizedImage,text,(50,150),font,1.5,(0,0,255),2)
cv2.imshow("Text on Image",resizedImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

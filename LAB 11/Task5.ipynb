```python
import cv2
image = cv2.imread("image2.png")
ret, thresholdedImage = cv2.threshold(image,150,255,cv2.THRESH_BINARY)
cv2.imshow("Thresholded Image",thresholdedImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rotationMatrix = cv2.getRotationMatrix2D((image.shape[1]/2,image.shape[0]/2),60,1)
rotatedImage = cv2.warpAffine(image,rotationMatrix,(image.shape[1],image.shape[0]))
cv2.imshow("Rotated Image",rotatedImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
image = cv2.imread("image.png")
image2 = cv2.imread("image2.png") 
image = cv2.resize(image,(image2.shape[1],image2.shape[0])) 
blended_image = cv2.add(image,image2)
cv2.imshow("Added Images",blended_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.cvtColor(blended_image,cv2.COLOR_BGR2GRAY)
cv2.imshow("Greyscale Image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
equalizedImage = cv2.equalizeHist(image)
cv2.imshow("Equalized Image",equalizedImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

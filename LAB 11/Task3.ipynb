```python
import cv2
image = cv2.imread("image.png")
blurredImage = cv2.GaussianBlur(image,(5,5),0)
cv2.imshow("Blurred Image",blurredImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
roi = image[100:600,150:750]
cv2.imshow("Cropped Image",roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
